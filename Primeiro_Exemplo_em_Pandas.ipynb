```python
revendas = combustiveis_df["Revenda"]
display(revendas)
```

```python
produtos_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(produtos_df)
```

```python
#Localiza uma linha (usando o índice, começa em zero)
#Quarta linha do dataset
```

```python
display(combustiveis_df.loc[3])
```

```python
#Terceira até a sexta linha
```

```python
display(combustiveis_df.loc[2:5])
```

```python
#localizar linhas que correspondam a uma condição específica
```

```python
vendas_gasolina_df = combustiveis_df.loc[combustiveis_df['Produto'] =='GASOLINA']
display(vendas_gasolina_df)

display(vendas_gasolina_df['Valor de Venda'].max())
```

```python
#Exibir qial é o posto e município da gasolina mais cara do país?
```

```python
display(combustiveis_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
#Múltiplas condições
```

```python
vendas_gasolina_df = combustiveis_df.loc[(combustiveis_df['Produto'] == 'GASOLINA') & (combustiveis_df['Municipio'] == 'INDAIATUBA')]
display(vendas_gasolina_df)
```

```python
#Condicional especificando Colunas
```

```python
vendas_gasolina_df = combustiveis_df.loc[(combustiveis_df['Produto'] == 'GASOLINA') & (combustiveis_df['Municipio'] == 'INDAIATUBA'), 
                                         ['Municipio', 'Revenda', 'Data da Coleta', 'Produto', 'Valor de Venda']]
display(vendas_gasolina_df)
```

```python
#Informação nova partir de uma coluna que existe
```

```python
vendas_gasolina_df['Imposto'] = vendas_gasolina_df['Valor de Venda'] * 0.10
display(vendas_gasolina_df)
```

```python
#Removendo colunas - axis=1 é coluna/axis=0 é linha
```

```python
vendas_gasolina_df = vendas_gasolina_df.drop("Imposto", axis=1)
display(vendas_gasolina_df)
```

```python
#Deletar linhas e colunas completamente vazias
```

```python
vendas_gasolina_df = vendas_gasolina_df.dropna(how='all')
display(vendas_gasolina_df)
```

```python
#Deletar colunas completamente vazias
```

```python
vendas_gasolina_df = vendas_gasolina_df.dropna(how='all', axis=1)
display(vendas_gasolina_df)
```

```python
#Preencher automático com a média da coluna
```

```python
vendas_gasolina_df['Imposto'] = vendas_gasolina_df['Valor de Venda'] * 0.10
vendas_gasolina_df['Imposto'] = vendas_gasolina_df['Imposto'].fillna(vendas_gasolina_df['Imposto'].mean())
display(vendas_gasolina_df)
```

```python
#Grupo by
```

```python
media_por_produto = combustiveis_df[['Produto', 'Valor de Venda']].groupby('Produto').mean()
display(media_por_produto)
```

```python
#Salvando um novo excel
```

```python
media_por_produto.to_excel("media_por_produto.xlsx")
```
