```python
#Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO?
```

```python
#Como mostrar média de valor de vendas Por Combustível?
```

```python
media_por_combustivel = combustiveis_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel)
```

```python
#E se eu quiser uma coluna nova chamada "Obs"
#e que preencha automaticamente "Melhor Cidade"
#quando o |Município for SAO PAULO, MG
```

```python
combustiveis_df['Obs'] = ["MELHOR CIDADE" if municipio == 'SAO PAULO' else None for municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO','INDAIATUBA', 'CAMPINAS', 'SALTO']), ['Municipio', 'Obs']])
```

```python
import numpy as np
combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display(combustiveis_df[['Revenda', 'Valor de Venda', 'Valor de Venda - Status']])
```

```python
#Calcular os postos de gasolina por habitantes temos na amostragem de combustíveis nov/2021
```

```python
num_habitantes_df = pd.read_csv("/content/ibge_num_habitantes_estimado.csv", sep=";")
num_habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
# Faz um MERGE dos dois dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
#Destruir coluna completamente vazia (todas as linhas são nulas)
merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
display(merge_df.head(100))
```

```python
merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
#Agrupar e contar quantos postos tem na cidade
postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "Número de Postos"}, inplace=True)
display(postos_por_municipio_df)
```

```python
#Múltiplas condições
```

```python
vendas_gasolina_df = combustiveis_df.loc[(combustiveis_df['Produto'] == 'GASOLINA') & (combustiveis_df['Municipio'] == 'BELO HORIZONTE')]
display(vendas_gasolina_df)
```

```python
#Condicional especificando Colunas
```

```python
vendas_gasolina_df = combustiveis_df.loc[(combustiveis_df['Produto'] == 'GASOLINA') & (combustiveis_df['Municipio'] == 'BELO HORIZONTE'), 
                                         ['Municipio', 'Revenda', 'Data da Coleta', 'Produto', 'Valor de Venda']]
display(vendas_gasolina_df)
```

```python
#Informação nova partir de uma coluna que existe
```

```python
vendas_gasolina_df['Imposto'] = vendas_gasolina_df['Valor de Venda'] * 0.10
display(vendas_gasolina_df)
```

```python
#Removendo colunas - axis=1 é coluna/axis=0 é linha
```

```python
vendas_gasolina_df = vendas_gasolina_df.drop("Imposto", axis=1)
display(vendas_gasolina_df)
```

```python
#Deletar linhas e colunas completamente vazias
```

```python
vendas_gasolina_df = vendas_gasolina_df.dropna(how='all')
display(vendas_gasolina_df)
```

```python
#Deletar colunas completamente vazias
```

```python
vendas_gasolina_df = vendas_gasolina_df.dropna(how='all', axis=1)
display(vendas_gasolina_df)
```

```python
#Preencher automático com a média da coluna
```

```python
vendas_gasolina_df['Imposto'] = vendas_gasolina_df['Valor de Venda'] * 0.10
vendas_gasolina_df['Imposto'] = vendas_gasolina_df['Imposto'].fillna(vendas_gasolina_df['Imposto'].mean())
display(vendas_gasolina_df)
```

```python
media_por_produto = combustiveis_df[['Produto', 'Valor de Venda']].groupby('Produto').mean()
display(media_por_produto)
```
